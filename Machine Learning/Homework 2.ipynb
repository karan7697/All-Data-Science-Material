```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv (r'multiTimeline.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Month'] =  pd.to_datetime(df['Month'])
```

```python
df.set_index('Month',inplace=True)
```

```python
df.head()
```

## General Visualisation of the graph:

```python
plt.figure(figsize=(12,6))
plt.plot(df.Diet, linewidth=1)
plt.plot(df.Gym, linewidth=1)
plt.plot(df.Finance,linewidth=1)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('General Visualization Trends:')
plt.legend(['Diet', 'Gym', 'Finance'])
plt.show()
```

## Diet Rolling Average Visualisation:

```python
plt.figure(figsize=(12,6))
plt.plot(df.Diet.rolling(15).mean())
plt.xlabel('Year')
plt.ylabel('People')
plt.legend(labels=['Diet'])
plt.title('Diet Visualisation:')
plt.show()
```

## Gym Rolling Average Visualisation:

```python
plt.figure(figsize=(12,6))
plt.plot(df.Gym.rolling(15).mean())
plt.xlabel('Year')
plt.ylabel('People')
plt.legend(labels=['Gym'])
plt.title('Gym Visualisation:')
plt.show()
```

## Compasrison of Gym and Diet:

```python
plt.figure(figsize=(12,6))
plt.plot(df.Gym.rolling(15).mean())
plt.plot(df.Diet.rolling(15).mean())
plt.xlabel('Year')
plt.ylabel('People')
plt.legend(labels=['Gym','Diet'])
plt.title('GYM AND DIET COMPASRISON VISUALISATION:')
plt.show()
```

## First OrderDifferencing: 

```python
diet = df[['Diet']]
diet.diff().plot(figsize=(12,6), linewidth=1)
plt.xlabel('Year');
```

## Auto-corelation Plot:

```python
pd.plotting.autocorrelation_plot(df['Diet']);
```

The statistical significance of the correlation is indicated by the dotted lines in the accompanying plot. In this scenario, the 'diet' series can be said to be legitimately autocorrelated with a twelve-month lag.
