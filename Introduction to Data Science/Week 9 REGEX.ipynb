Create start date, end date columns and populate the values (If notes don't include month or day use 1) i.e: 2018 => 2018-01-01

Create phone number column (If notes don't include phone numbers, use None)

Create a column duration showing difference between start & end dates in years as a whole number

```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week09/data/hw_data.csv', index_col=0)
```

```python
df
```

```python
#Extracting Phone Numbers from the dataset
pattern = re.compile(r'(\d{3})?((\(\d{3}\))|(\d{3})(\s|-?|.?))(\d{3}(\s|-?|.?))(\d{4})(\s?(([E|e]xt[:|.|]?))(\s?\d+))?')
```

```python
df['PhNo']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
df.head(20)
```

```python
#Pattern To Seperate Start Date
pattern = re.compile(r'((start|st\sdt|st\sdate|st\sDt|STDT)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4})',re.I)
```

```python
df['StartDate']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
```

```python
#Pattern to Extract End Date
pattern = re.compile(r'(end|ed|en\sdt|endt)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4}',re.I)
```

```python
df['EndDate']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
```

```python
df
```

```python
df['stdt'] = df['StartDate'].str.replace(r'(start|st\sdt|st\sdate|st\sDt|STDT)\D*', '')
```

```python
# df['enddate'] =  df['EndDate'].str.replace(r'(\Wend\sdate|ed\sdt|/ENDT|-ENDT|ENDT|end|ed|en\sdt)\D*', '')
df['ED'] =  df['EndDate'].str.replace(r'(end|ed|en\sdt|ENDT)\D*', '')
```

```python
df.head(50)
```

```python
df.head(50)
```

```python
# pattern = re.compile(r'(\b^[0-9]{2}(\/)[0-9]{2})$\b')
pattern = re.compile(r'(\b^[0-9]{2}(\/)[0-9]{2})$\b')
```

```python
df['start']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['stdt']]
```

```python
df['end']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['ED']]
```

```python
df.head(50)
```

```python
df['start_year']=pd.DatetimeIndex(df['start']).year
df['end_year']=pd.DatetimeIndex(df['end']).year
```

```python
#droping unecessary columns
df.drop(['StartDate','EndDate','stdt','ED','notes'], axis=1, inplace=True)
df.head(3)
```

```python
#calculating the duration in years
df['Duration']=df['end_year']-df['start_year']
```
